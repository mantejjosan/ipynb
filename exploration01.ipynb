```python
import matplotlib.pyplot as plt
import math

# Input lengths
base = float(input("Enter base length: "))
height = float(input("Enter height: "))

# Calculate hypotenuse
hypotenuse = math.sqrt(base**2 + height**2)
print(f"Hypotenuse: {hypotenuse}")

# Plotting the triangle
plt.figure(figsize=(6, 6))
plt.plot([0, base, base, 0], [0, 0, height, 0], marker='o')

# Annotating the sides
plt.text(base / 2, -0.5, f"Base: {base}", ha='center')
plt.text(base + 0.2, height / 2, f"Height: {height}", va='center')
plt.text(base / 2, height / 2, f"Hypotenuse: {round(hypotenuse, 2)}", ha='center', rotation=math.degrees(math.atan(height/base)))

# Formatting the plot
plt.title("Right-Angled Triangle")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.axis('equal')  # Keep proportions true to scale
plt.show()
```

```python
import matplotlib.pyplot as plt
import cmath  # For complex math operations

# Input complex numbers
base = complex(input("Enter base (as a complex number, e.g., 3+4j): "))
height = complex(input("Enter height (as a complex number, e.g., 1+2j): "))

# Calculate hypotenuse as the modulus of their sum
hypotenuse = abs(base + height)
print(f"Hypotenuse: {hypotenuse}")

# Extract real and imaginary parts
x_values = [0, base.real, (base + height).real, 0]
y_values = [0, base.imag, (base + height).imag, 0]

# Plot the triangle in the complex plane
plt.figure(figsize=(6, 6))
plt.plot(x_values, y_values, marker='o', linestyle='-', label="Triangle")

# Annotate points
plt.text(base.real / 2, base.imag / 2, "Base", ha='center', fontsize=10)
plt.text((base + height).real / 2, (base + height).imag / 2, "Hypotenuse", ha='center', fontsize=10)
plt.text(base.real + 0.2, base.imag, "Height", fontsize=10)

# Formatting the plot
plt.axhline(0, color='black', linewidth=0.5)  # X-axis
plt.axvline(0, color='black', linewidth=0.5)  # Y-axis
plt.grid(True)
plt.xlabel("Real Part")
plt.ylabel("Imaginary Part")
plt.title("Triangle in the Complex Plane")
plt.axis("equal")
plt.legend()
plt.show()
```
